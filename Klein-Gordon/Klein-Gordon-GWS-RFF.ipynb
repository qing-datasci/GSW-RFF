```python
import torch
import numpy as np
import torch.nn as nn
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from torch.autograd import grad
from pyDOE import lhs
import random
from mpl_toolkits.axes_grid1 import make_axes_locatable
import sys 
import matplotlib.ticker as ticker


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
dtype=torch.float32
torch.set_default_dtype(dtype)


def setup_seed(seed):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cuda.matmul.allow_tf32 = False
        torch.backends.cudnn.allow_tf32 = False
        #torch.backends.cudnn.deterministic = True
        #torch.backends.cudnn.benchmark = False


seed = 8888

setup_seed(seed)


class layer(nn.Module):
    def __init__(self, n_in, n_out, activation):
        super().__init__()
        self.layer = nn.Linear(n_in, n_out)
        self.activation = activation

    def forward(self, x):
        x = self.layer(x)
        if self.activation:
            x = self.activation(x)
        return x



class DNN(nn.Module):
    def __init__(self, dim_in, dim_out, n_layer, n_node, ub, lb, activation=nn.Tanh()):
        super().__init__()
        self.net0 = nn.ModuleList()
        self.net0.append(layer(n_node, n_node, activation))
        for _ in range(n_layer):
            self.net0.append(layer(n_node, n_node, activation))


        self.ub = torch.tensor(ub, dtype=torch.float).to(device)
        self.lb = torch.tensor(lb, dtype=torch.float).to(device)
        self.net0.apply(weights_init)  # xavier initialization

        
        self.fc2 = nn.Linear(2*n_node, dim_out)
        self.A0 = 1*torch.randn((dim_in,int(n_node/2))).to(device)
        self.A1 = 5*torch.randn((dim_in,int(n_node/2))).to(device)

    def forward(self, x):
        out = x
        out0 = out@self.A0
        out00 = torch.sin(out0)
        out01 = torch.cos(out0)
        out0 = torch.cat((out00,out01),1)


        out1 = out@self.A1
        out10 = torch.sin(out1)
        out11 = torch.cos(out1)
        out1 = torch.cat((out10,out11),1)
        for layer in self.net0:
            out0 = layer(out0)
            out1 = layer(out1)

            
        out = torch.cat((out0,out1),1)
        out = self.fc2(out)
        return out



def weights_init(m):
    if isinstance(m, nn.Linear):
        torch.nn.init.xavier_uniform_(m.weight.data)
        torch.nn.init.zeros_(m.bias.data)

def rhs(x):
    
    return -6*(x[:,0:1]**3)*x[:,1:2]+x[:,1:2]*(6*x[:,0:1]*x[:,1:2]**2-25*(np.pi**2)*torch.cos(5*np.pi*x[:,0:1]))+((x[:,0:1]*x[:,1:2])**3+x[:,1:2]*torch.cos(5*np.pi*x[:,0:1]))**3
    


def u(t,x):
    Z = x*np.cos(5*np.pi*t)+(x*t)**3
    
    return Z


# Parameters
t_min = 0
t_max = 1
x_min = 0
x_max = 1


ub = np.array([t_max, x_max])
lb = np.array([t_min, x_min])


N_bc = 500
N_f = 5000




t_points = lambda n: np.random.uniform(t_min, t_max, (n, 1))
x_points = lambda n: np.random.uniform(x_min, x_max, (n, 1))








class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        
        self.net = DNN(dim_in=2, dim_out=1, n_layer=3, n_node=60, ub=ub, lb=lb).to(device)#,activation=torch.nn.GELU()
        self.iter = 0
        self.optimizer = torch.optim.LBFGS(
            self.net.parameters(),
            lr=1,
            max_iter=20000,
            max_eval=20000,
            history_size=200,
            tolerance_grad=1e-12, 
            tolerance_change=1.0 * np.finfo(float).eps,
            line_search_fn="strong_wolfe",
        )#tolerance_change=1.0 * np.finfo(float).eps,
        self.adam = torch.optim.Adam(self.net.parameters(), lr=2e-3)
        self.scheduler=torch.optim.lr_scheduler.StepLR(self.adam,step_size=30,gamma=0.99)
        self.losses = {"loss": [], "bc": [], "pde": []}
       
    
    def loss_bc(self, xyt,u_t):
        u = self.net(xyt)
        mse_ic = torch.mean(torch.square(u - u_t))
        return mse_ic
    
    def loss_bc1(self, xyt):
        xy = xyt.clone()
        xy.requires_grad = True
        
        u = self.net(xy)
        u_t = grad(u, xy, grad_outputs=torch.ones_like(u),create_graph=True)[0][:, 0:1]
       
        mse_ic = torch.mean(torch.square(u_t))
        return mse_ic

    def loss_pde(self, xyt):
        xy = xyt.clone()
        xy.requires_grad = True
        
        u = self.net(xy)
        

        u_xyt = grad(u, xy, grad_outputs=torch.ones_like(u),create_graph=True)[0]
        u_xx = grad(u_xyt[:, 1:2].sum(), xy,create_graph=True)[0][:, 1:2]
        u_tt = grad(u_xyt[:, 0:1].sum(), xy,create_graph=True)[0][:, 0:1]

        
        pde =-rhs(xy)+u_tt+u**3 - u_xx
        
        
        return pde
    
    

    def closure(self):
        temp=int(N_f/m/n)
        self.optimizer.zero_grad()
        self.adam.zero_grad()
        
        
        mse_bc = self.loss_bc(tx_bc,u_bc)
        mse_ic = self.loss_bc(tx_ic,u_ic)
        mse_bc1 = self.loss_bc1(tx_ic)
        pde = self.loss_pde(tx_f)
        mse_pde=0
        for i in range(m*n):
            
            temp0=(1/m/n)*torch.mean(torch.square(pde[(i*temp):((i+1)*temp)]))
            mse_pde+=omega[i]*temp0

        loss = (mse_ic+ mse_bc) + mse_pde +mse_bc1
        
        loss.backward()

        self.losses["loss"]=loss.detach().cpu().item()
        self.losses["ic"]=mse_ic.detach().cpu().item()
        self.losses["bc"]=mse_bc.detach().cpu().item()
        self.losses["bc1"]=mse_bc1.detach().cpu().item()
        self.losses["pde"]=mse_pde.detach().cpu().item()
        
        
        if self.iter % 1000 == 0:
        
            print(f"{self.iter} Loss: {loss.item():.5e}  IC: {mse_ic.item():.5e} BC: {mse_bc.item():.5e} BC1: {mse_bc1.item():.5e} pde:{mse_pde.item():.5e}")
        self.iter += 1
        return loss


    
    
#log_t='log1_normal.txt'    
if __name__ == "__main__":
    m=1;n=1

    point=np.empty((0,2))
    omega=np.zeros(m*n)

    pinn = PINN()
    
    
    
    
    t_bc1 = np.ones((N_bc, 1))*t_min
    x_bc1 = x_points(N_bc)

    

    t_bc3 = t_points(N_bc//2)
    x_bc3 = np.ones((N_bc//2, 1))*x_min

    
    t_bc4 = t_points(N_bc//2)
    x_bc4 = np.ones((N_bc//2, 1)) * x_max


    t_bc = np.vstack([t_bc3, t_bc4])
    x_bc = np.vstack([x_bc3, x_bc4])
    tx_bc = np.hstack([t_bc, x_bc])
    tx_ic = np.hstack([t_bc1,x_bc1])
    u_bc = u(t_bc,x_bc)
    u_ic = u(t_bc1,x_bc1)
    
    temp_t = np.vstack([t_bc, t_bc1])
    temp_x = np.vstack([x_bc, x_bc1])
    u_temp = u(temp_t,temp_x)
    omega1 = np.sqrt(len(temp_t))/np.linalg.norm(u_temp, ord=2)
    
    for i in range(m):
        for j in range(n):
            t_f = (t_min+j*((t_max-t_min)/n)) + ((t_max-t_min)/n) * lhs(1, int(N_f/m/n))
            x_f = (x_min+i*((x_max-x_min)/m)) + ((x_max-x_min)/m) * lhs(1, int(N_f/m/n))
            xy_f = np.hstack([t_f, x_f])
            point=np.vstack([point, xy_f])
            temp= torch.tensor(xy_f, dtype=torch.float32,device=device)
            norm_f=rhs(temp)
            omega[i*n+j]=np.sqrt(len(x_f))/((torch.linalg.vector_norm(norm_f,2).item()))#+len(x_f))

    tx_bc = torch.tensor(tx_bc, dtype=dtype).to(device)
    tx_ic = torch.tensor(tx_ic, dtype=dtype).to(device)
    u_ic = torch.tensor(u_ic, dtype=dtype).to(device)
    u_bc = torch.tensor(u_bc, dtype=dtype).to(device)
    tx_f = torch.tensor(point, dtype=dtype).to(device)
   
```

```python
def cm2inch(*tupl):
    '''
    Specify figure size in centimeter in matplotlib
    Source: http://stackoverflow.com/a/22787457/395857
    By gns-ank
    '''
    inch = 2.54
    if type(tupl[0]) == tuple:
        return tuple(i/inch for i in tupl[0])
    else:
        return tuple(i/inch for i in tupl)

def heatmap(AUC, title, xlabel, ylabel, xticklabels, yticklabels):
    '''
    Inspired by:
    - http://stackoverflow.com/a/16124677/395857
    - http://stackoverflow.com/a/25074150/395857
    '''

    # Plot it out
    fig, ax = plt.subplots()    
    c = ax.pcolor(AUC, edgecolors='k', linestyle= 'dashed', linewidths=0.2, cmap='YlOrRd')

    # put the major ticks at the middle of each cell
    ax.set_yticks(np.arange(AUC.shape[0]) + 0.5, minor=False)
    ax.set_xticks(np.arange(AUC.shape[1]) + 0.5, minor=False)

    # set tick labels
    #ax.set_xticklabels(np.arange(1,AUC.shape[1]+1), minor=False)
    ax.set_xticklabels(xticklabels, minor=False)
    ax.set_yticklabels(yticklabels, minor=False)

    # set title and x/y labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)      

    # Remove last blank column
    plt.xlim( (0, AUC.shape[1]) )

    # Turn off all the ticks
    ax = plt.gca()    
    for t in ax.xaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
    for t in ax.yaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False

    # Add color bar
    plt.colorbar(c)


    # resize
    fig = plt.gcf()
    fig.set_size_inches(cm2inch(40, 20))    
```

```python
error_vec_inf_2 = []
error_vec_2_2 = []
for ttt in range(1):10
    
    
    
    t_bc1 = np.ones((N_bc, 1))*t_min
    x_bc1 = x_points(N_bc)

    

    t_bc3 = t_points(N_bc)
    x_bc3 = np.ones((N_bc, 1))*x_min

    
    t_bc4 = t_points(N_bc)
    x_bc4 = np.ones((N_bc, 1)) * x_max


    t_bc = np.vstack([t_bc3, t_bc4])
    x_bc = np.vstack([x_bc3, x_bc4])
    tx_bc = np.hstack([t_bc, x_bc])
    tx_ic = np.hstack([t_bc1,x_bc1])
    u_bc = u(t_bc,x_bc)
    u_ic = u(t_bc1,x_bc1)
    
    
    
    for i in range(m):
        for j in range(n):
            t_f = (t_min+j*((t_max-t_min)/n)) + ((t_max-t_min)/n) * lhs(1, int(N_f/m/n))
            x_f = (x_min+i*((x_max-x_min)/m)) + ((x_max-x_min)/m) * lhs(1, int(N_f/m/n))
            xy_f = np.hstack([t_f, x_f])
            point=np.vstack([point, xy_f])
            temp= torch.tensor(xy_f, dtype=torch.float32,device=device)
            norm_f=rhs(temp)
            #norm_t=norm_f-torch.min(norm_f)
            #by[i*m+j]=(torch.std(norm_t).item())/(torch.mean(norm_t).item()+1)
            omega[i*n+j]=np.sqrt(len(x_f))/((torch.linalg.vector_norm(norm_f,2).item()))#+len(x_f))
            

    
    
    tx_bc = torch.tensor(tx_bc, dtype=dtype).to(device)
    tx_ic = torch.tensor(tx_ic, dtype=dtype).to(device)
    u_ic = torch.tensor(u_ic, dtype=dtype).to(device)
    u_bc = torch.tensor(u_bc, dtype=dtype).to(device)
    tx_f = torch.tensor(point, dtype=dtype).to(device)
    
    #print(j)
    pinn = PINN()
    print("第",ttt,"次训练——————————————————————————————")
    
    for k in range(5000):#4000
        pinn.closure()
        pinn.adam.step()
        pinn.scheduler.step()
    pinn.optimizer.step(pinn.closure)
    print(f"{pinn.iter} Loss: {pinn.losses['loss']:.5e}  IC: {pinn.losses['ic']:.5e} BC: {pinn.losses['bc']:.5e} BC1: {pinn.losses['bc1']:.5e} pde: {pinn.losses['pde']:.5e}")
    T0,X0 = np.meshgrid(np.linspace(0,1,101),np.linspace(0,1,101))
    Z0 = X0*np.cos(5*np.pi*T0)+(X0*T0)**3
    tX,tY = torch.tensor(T0,dtype=torch.float32,device=device),torch.tensor(X0,dtype=torch.float32,device=device)
    with torch.inference_mode(1):tZ = pinn.net(torch.cat([tX.reshape(-1,1),tY.reshape(-1,1)],dim=1)).reshape(101,101)
    U0 = tZ.cpu().numpy()
    err=torch.from_numpy(Z0-U0)
    u_norm=torch.from_numpy(U0)
    error_vec_inf_2.append(float( torch.linalg.vector_norm(err,float('inf')).numpy()))
    error_vec_2_2.append(float((torch.linalg.vector_norm(err,2)/torch.linalg.vector_norm(u_norm,2)).numpy()))
#torch.save(pinn.state_dict(), 'KG-rff.pt')    
```
